```python
class Talaba:
    def __init__(self,name,sec_name,b_year):
        self.name=name
        self.sec_name=sec_name
        self.b_year=b_year
        self.kurs=2
    
    def get_name(self):
        return self.name
    
    def get_sec_name(self):
        return self.sec_name
    
    def get_age(self,year):
        return year-self.b_year
    
    def set_bosqich(self,yangi_kurs):
        self.kurs=yangi_kurs
        #return yangi_kurs
    
    def update_bosqich(self):
        self.kurs+=1
        
    def get_info(self):
        return f"{self.name} {self.sec_name}. {self.kurs} - kurs talabasi"
    
talaba1=Talaba('Dilshod','Atadjanov',1989)
talaba2=Talaba('Anvar','Atadjanov',1980)
talaba3=Talaba('Alisher','Atadjanov',1983)
#print(talaba1.name, " ",talaba1.kurs,"- kursda uqiydi")
#print(f'Ismi:{talaba1.get_name()} yoshi:{talaba1.get_age(2022)} da')
```
