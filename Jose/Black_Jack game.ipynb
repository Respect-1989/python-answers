```python
from IPython.display import clear_output
clear_output()
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
class Card:
    clear_output()
    def __init__(self,rank,suit):
        self.rank=rank
        self.suit=suit
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Desk:
    def __init__(self):
        self.desk=[]
        for suit in suits:
            for rank in ranks:
                self.desk.append(Card(rank,suit).__str__())
                
    def __str__(self):
        card_desk=''
        for i in self.desk:
            card_desk+='\n'+i
        return "The cards on the desk: "+card_desk
    
    def shuffle(self):
        random.shuffle(self.desk)

    def pop_card(self):
        return self.desk.pop()   
desk=Desk()
```

```python
class Hand:
    def __init__(self):
        self.d_value=0
        self.p_value=0

    def get_dealer(self,dealer):
        a1=values[str(dealer).split()[0]]
        self.d_value+=a1
        return f"{self.d_value}"
    
    def get_player(self,player):
        a1=values[str(player).split()[0]]
        self.p_value+=a1
        return f"{self.p_value}"
```

```python
h=Hand()
class Players:
    def __init__(self):
        self.dealer_hand=[]
        self.player_hand=[]
        self.d_value=0
        self.p_value=0
        
    def deal(self,d_card):
        self.dealer_hand.append(d_card)
        self.d_value=h.get_dealer(d_card)
        return self.dealer_hand
    
    def dealer_value(self):
        return self.d_value
        
    def play(self,p_card):
        self.player_hand.append(p_card)
        self.p_value=h.get_player(p_card)
        return self.player_hand
    
    def player_value(self):
        return self.p_value
```

```python
class Cards:
    def __init__(self):
        self.dll=''
                
    def add_card(self,get_pop):
        self.dll=''
        for i in get_pop:
            self.dll+='\n'+i 
        return self.dll
```

```python
class Draw:
    def __init__(self,balance):
        self.balance=balance
        
    def add_balance(self,amount):
        self.balance+=amount
        return self.balance

    def get_withdraw(self,withdraw):
        if withdraw<=self.balance:
            self.balance-=withdraw
        else:
            print(f'Sorry, you don\'t have enough chips! You have: {self.balance}')
            
    def winning(self,dealer,player,withdraw):
        
        if dealer>player and dealer<=21:
            self.balance-=withdraw
            print(f'Player\'s winnings stand at {self.balance}')
        elif dealer<player and player<=21:
            self.balance+=withdraw  
            print(f'Player\'s winnings stand at {self.balance}')
        elif dealer<player and player>21:
            self.balance-=withdraw  
            print(f'Player\'s winnings stand at {self.balance}')
        elif dealer>player and dealer>21:
            self.balance+=withdraw  
            print(f'Player\'s winnings stand at {self.balance}')
        else:
            print('!!!')
```
