```python
import random

def get_word(words):
    word=random.choices(words)
    while "-" in word or " " in word:
        word=random.choices(words)
    return word
word=get_word(words)


def display(word):
    a=list(''.join(word)) # ['а', 'г', 'р', 'е', 'с', 'с', 'о', 'р'] 
    b=list('-'*len(a)) #-----------
    disp_word=''.join(word)
    print(f'a: {a}, word: {word} disp_word: {disp_word} b:{b}')
    display_letter=""
    while word:
        letter=input()
        if letter in disp_word:
            c=a.index(letter) # index is a 2,5,or 7 
            display_letter+=letter
            #b[c]+=letter ####<--------------------------
            #b[c]=a
            #a.remove(letter)
            if b[c]=='-':
                b[c]=a[c] #b[2]=a[2] 
                #a.remove(letter)#remove a[2]
            if b[c]!='-':
                b[c+1]
            print(word, display_letter,b,a)
            
        else:
            print('sorry')
display(word)
```
