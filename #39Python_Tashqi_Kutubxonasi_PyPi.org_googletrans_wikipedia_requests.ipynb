# googletrans


```python
#pip install googletrans==3.1.0a0
from googletrans import Translator
tarjimon=Translator()
matn_uz="Python eng yaxshi til"
tarjima=tarjimon.translate(matn_uz)
if tarjima.text==matn_uz:
    print('So\'zni xatosini tekshiring!')
else:
    print(tarjima.origin)
    print(tarjima.text)
    print(tarjima.src)
    
```

```python
matn_uz="Python eng yaxshi til"
tarjima=tarjimon.translate(matn_uz,dest='ru')
print(tarjima.origin)
print(tarjima.text)
print(tarjima.src)
```

```python
matn_en="Python is very best language in the world"
tarjima=tarjimon.translate(matn_en,dest='uz')
print(tarjima.origin)
print(tarjima.text)
print(tarjima.src)
```

```python
from googletrans import Translator

tarjimon=Translator()
msg='Tarjima uchun so\'z kirgizing.'
while True:
    texts=input(msg)
    if texts=='q':
        break
    else:
        tarjima=tarjimon.translate(texts, src='uz', dest='en')
        print(tarjima.origin)
        print(tarjima.text)
```

```python
import requests
from googletrans import Translator
from pprint import pprint

url="https://api.adviceslip.com/advice"
r=requests.get(url)
advice=r.json()['slip']['advice']
print(advice)

tarjimon=Translator()
tarjima=tarjimon.translate(advice,dest='uz')
perevod=tarjimon.translate(advice,dest='ru')
ar=tarjimon.translate(advice,dest='de')
print(perevod.text)
print(tarjima.text)
print(ar.text)
```
